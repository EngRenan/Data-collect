# Coletando dados estruturados em lista de arquivos

Neste arquivo mostrarei como coletar dados através de varios arquivos separados por datas. O dataset utilizado é ficticio que ilustra preferencias de alguns entrevistados.

data.PNG

A pratica consiste em coletar todos os dados em cada arquivo e adiciona-los em um dataframe do pandas e em seguida analisar e persistir os dados em arquivo de texto e em tabela no banco de dados. 

```python
# Importando bibliotecas 
import pandas as pd
import os # biblioteca para funções do sistema operacional
```

```python
# Atribuindo variavel caminho no qual estão os arquivos 
diretorio = '/content/drive/MyDrive/BASE_TP'
```

Agora vamos criar uma lista com todos os arquivos que estao armazenados no diretorio  utilizando o "listdir" no qual retornará uma lista contendo os nomes de entrada no diretorio fornecido por path.

```python
lista_df = [] # lista vazia
lista_path = os.listdir(diretorio)
lista_path[:5]
```

```python
# Com os nomes dos arquivos que estão dentro do diretorio, vamos abrir cada arquivo e coletar os dados
lista_df = [] #lista vazia
for arquivo in lista_path:
  local_arquivo = os.path.join(diretorio, arquivo)
  df = pd.read_csv(local_arquivo, sep='|', encoding='latin-1')
  lista_df.append(df)

df_entrevista = pd.concat(lista_df) # atribui os dados coletados em um dataframe
```

```python
df_entrevista[:10]
```

# Exportando dados para arquivos

```python
# Exportando dados
df_entrevista.to_csv(diretorio + 'entrevista_geral', sep = '|', encoding='latin-1', index=False)
```
